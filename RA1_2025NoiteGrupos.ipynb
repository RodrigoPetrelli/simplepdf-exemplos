Considerando um fenômeno que possa ser explicado pela equação

Encontre o coeficiente angular da tangente a dois pontos que você ache que sejam relevantes para o entendimento deste fenômeno. # **1. QUESTÃO 1**

## **1.1 Introdução**
O coeficiente angular da reta tangente pode ser aproximado pela taxa média de variação, dada pela fórmula:


$$m = \frac{y_2 - y_1}{x_2 - x_1}$$

Para a função:

$$y = 3x^3 - x^2 + \sin(10x)$$

Escolhemos dois pares de pontos para calcular a inclinação da reta secante.

---

## **1.2 Cálculo dos Coeficientes Angulares**

### **Primeiro intervalo: $x_1 = 0$ e $x_2 = 1$**  
Calculamos os valores da função:

$$y_1 = f(0) = 3(0)^3 - (0)^2 + \sin(10 \cdot 0) = 0$$

$$y_2 = f(1) = 3(1)^3 - (1)^2 + \sin(10 \cdot 1)$$

Aproximamos:

$$y_2 \approx 3(1) - 1 + (-0.544) = 3 - 1 - 0.544 = 1.456$$

Agora, calculamos a taxa média de variação:

$$m_1 = \frac{y_2 - y_1}{x_2 - x_1} = \frac{1.456 - 0}{1 - 0} = 1.456$$

---

### **Segundo intervalo: $x_3 = 1$ e $x_4 = 2$**
Novamente, calculamos os valores da função:

$$y_3 = f(1) = 1.456$$

$$y_4 = f(2) = 3(2)^3 - (2)^2 + \sin(10 \cdot 2)$$

Aproximamos:

$$y_4 \approx 3(8) - 4 + (-0.544) = 24 - 4 - 0.544 = 19.456$$

Agora, calculamos a taxa média de variação:

$$m_2 = \frac{y_4 - y_3}{x_4 - x_3} = \frac{19.456 - 1.456}{2 - 1} = 18$$
---

## **1.5 Conclusão**  
O polinômio $3x^3 - x^2$ dita uma tendência de crescimento, enquanto $\sin(10x)$ apresenta variações rápidas e oscilatórias.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definição da função f(x)
def f(x):
    return 3 * x**3 - x**2 + np.sin(10 * x)

# Pontos de interesse
x1, x2 = 0, 1
x3, x4 = 1, 2

# Cálculo dos valores da função
y1 = f(x1)
y2 = f(x2)
y3 = y2  # y3 = f(1) já calculado
y4 = f(x4)

# Cálculo das taxas médias de variação
m1 = (y2 - y1) / (x2 - x1)
m2 = (y4 - y3) / (x4 - x3)

# Plotando a função e as retas secantes
x_vals = np.linspace(-0.5, 2.5, 1000)
y_vals = f(x_vals)

# Retas secantes
sec1_x = np.array([x1, x2])
sec1_y = np.array([y1, y2])

sec2_x = np.array([x3, x4])
sec2_y = np.array([y3, y4])

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = 3x^3 - x^2 + \sin(10x)$', color='blue')
plt.plot(sec1_x, sec1_y, 'r--', label=f'Secante 1: m1 = {m1:.3f}')
plt.plot(sec2_x, sec2_y, 'g--', label=f'Secante 2: m2 = {m2:.3f}')

# Destacando os pontos
plt.scatter([x1, x2, x3, x4], [y1, y2, y3, y4], color='black', zorder=3)
for (x, y) in zip([x1, x2, x3, x4], [y1, y2, y3, y4]):
    plt.text(x, y, f'({x}, {y:.3f})', fontsize=10, verticalalignment='bottom')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.title("Gráfico da Função e das Retas Secantes")
plt.grid(True)
plt.show()
```

# **2. QUESTÃO 2**

## **2.1 Primeiro Limite**  
$$\lim_{x \to 2} \frac{x^2 - 5x + 6}{x^2 - 12x + 20}$$  

### **2.1.1 Fatoração dos Polinômios**  

O numerador pode ser fatorado como:  
$$x^2 - 5x + 6 = (x - 2)(x - 3)$$  

O denominador pode ser fatorado como:  
$$x^2 - 12x + 20 = (x - 2)(x - 10)$$  

### **2.1.2 Cancelamento de Termos**  

$$\frac{(x - 2)(x - 3)}{(x - 2)(x - 10)}$$  

Cancelamos o termo $ (x - 2) $, restando:  

$$\lim_{x \to 2} \frac{x - 3}{x - 10}$$  

Substituímos $ x = 2 $:  

$$\frac{2 - 3}{2 - 10} = \frac{-1}{-8} = \frac{1}{8}$$  

**Resposta:**  
$$\frac{1}{8}$$  

---

## **2.2 Segundo Limite**  
$$\lim_{x \to 3} \frac{x^4 - 81}{x^2 - 9}$$  

### **2.2.1 Fatoração**  

Utilizamos a diferença de quadrados:  
$$x^4 - 81 = (x^2 - 9)(x^2 + 9)$$  

Sabemos que:  
$$x^2 - 9 = (x - 3)(x + 3)$$  

Portanto, a fração se reescreve como:  
$$\frac{(x - 3)(x + 3)(x^2 + 9)}{(x - 3)(x + 3)}$$  

Cancelamos $ (x - 3)(x + 3) $, restando:  

$$\lim_{x \to 3} x^2 + 9$$  

Substituímos $ x = 3 $:  

$$3^2 + 9 = 9 + 9 = 18$$  

**Resposta:**  
$$18$$  

---

## **2.3 Terceiro Limite**  
$$\lim_{x \to 0} \left( \frac{1}{x} - \frac{1}{x^2} \right)$$  

### **2.3.1 Evidência**  

Colocamos $ \frac{1}{x} $ em evidência:  

$$\frac{1}{x} \left( 1 - \frac{1}{x} \right)$$  

Quando $ x \to 0 $, o termo $ \frac{1}{x} $ tende a $ \infty $ e $ \frac{1}{x^2} $ cresce ainda mais rápido para $ \infty $, resultando em um limite **infinito**.  

$$\lim_{x \to 0} \left( \frac{1}{x} - \frac{1}{x^2} \right) = -\infty$$  

**Resposta:**  
$$-\infty$$  

---

## **2.4 Quarto Limite**  
$$\lim_{x \to 0} \frac{\sqrt{1 + x} - 1}{-x}$$  

### **2.4.1 Multiplicação pelo Conjugado**  

Multiplicamos numerador e denominador pelo conjugado $ \sqrt{1 + x} + 1 $:  

$$\frac{(\sqrt{1 + x} - 1)(\sqrt{1 + x} + 1)}{-x (\sqrt{1 + x} + 1)}$$  

No numerador, aplicamos diferença de quadrados:  

$$(1 + x) - 1 = x$$  

Cancelamos $ x $, restando:  

$$\frac{1}{-(\sqrt{1 + x} + 1)}$$  

Quando $ x \to 0 $, temos $ \sqrt{1 + x} \to 1 $, logo:  

$$\frac{1}{-(1 + 1)} = \frac{1}{-2} = -\frac{1}{2}$$  

**Resposta:**  
$$-\frac{1}{2}$$  

---

## **2.5 Quinto Limite**  
$$\lim_{x \to 4} \frac{\sqrt{x^2 - 16}}{2x - 8}$$  

### **2.5.1 Fatoração**  

Fatoramos o denominador:  

$$2x - 8 = 2(x - 4)$$  

Reescrevemos o numerador como:  

$$\sqrt{x^2 - 16} = \sqrt{(x - 4)(x + 4)}$$  

### **2.5.2 Multiplicação pelo Conjugado**  

Multiplicamos numerador e denominador pelo conjugado $ \sqrt{x^2 - 16} + 4 $:  

$$\frac{(x - 4)(x + 4)}{2(x - 4)(\sqrt{x^2 - 16} + 4)}$$  

Cancelamos $ (x - 4) $, restando:  

$$\frac{x + 4}{2(\sqrt{x^2 - 16} + 4)}$$  

Substituímos $ x = 4 $:  

$$\frac{4 + 4}{2(0 + 4)} = \frac{8}{8} = 1$$  

**Resposta:**  
$$1$$  

---

## **2.6 Conclusão**  

Os valores dos limites calculados foram:  

- $$\lim_{x \to 2} \frac{x^2 - 5x + 6}{x^2 - 12x + 20} = \frac{1}{8}$$  

- $$\lim_{x \to 3} \frac{x^4 - 81}{x^2 - 9} = 18$$  

- $$\lim_{x \to 0} \left( \frac{1}{x} - \frac{1}{x^2} \right) = -\infty$$  

- $$\lim_{x \to 0} \frac{\sqrt{1 + x} - 1}{-x} = -\frac{1}{2}$$  

- $$\lim_{x \to 4} \frac{\sqrt{x^2 - 16}}{2x - 8} = 1$$
